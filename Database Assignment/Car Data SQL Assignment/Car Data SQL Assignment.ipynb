```python
import mysql.connector as conn
import csv
```

```python
cursor = sqlConn.cursor()
```

## 2 . create a database called as cardataset

```python
## Creating car_db
cursor.execute("CREATE DATABASE car_db")
```

## 3 . create a table called as car with a column name given in dataset description

```python
## Creating table1 inside the car_db

cursor.execute("USE car_db")

query = "CREATE TABLE car(buying VARCHAR(8), maint VARCHAR(8), doors VARCHAR(8), persons VARCHAR(8), lug_boot VARCHAR(8), safety VARCHAR(8), class VARCHAR(8))"
cursor.execute(query)
```

```python
cursor.execute("SHOW tables")
```

## 4 . Dump all the data into car table

```python
## Data Preview
with open('car.data', 'r') as f:
    car_data = csv.reader(f, delimiter = '\n')
    for i in car_data:
        row_data = i[0].split(',')
        print(row_data)
```

```python
# Dumping data
```

```python
with open('car.data', 'r') as f:
    car_data = csv.reader(f, delimiter = '\n')
    for i in car_data:
        row_data = i[0].split(',')
        query = "INSERT INTO car_db.car VALUES('{}','{}','{}','{}','{}','{}','{}')".format(str(row_data[0]), str(row_data[1]), str(row_data[2]), str(row_data[3]), str(row_data[4]), str(row_data[5]), str(row_data[6]))
        #query = "INSERT INTO car_db.car VALUES({})".format(str(i[0]))
        cursor.execute(query)
sqlConn.commit()
```

### Alternate Way to INSERT same data

```python
## Creating table1 inside the car_db

cursor.execute("USE car_db")

query = "CREATE TABLE car1(buying VARCHAR(8), maint VARCHAR(8), doors VARCHAR(8), persons VARCHAR(8), lug_boot VARCHAR(8), safety VARCHAR(8), class VARCHAR(8))"
cursor.execute(query)
```

```python
with open('car.data', 'r') as f:
    car_data = csv.reader(f, delimiter = '\n')
    for i in car_data:
        row_data = tuple(i[0].split(','))
        query = "INSERT INTO car_db.car1 VALUES {}".format(row_data)
        cursor.execute(query)
sqlConn.commit()
```

## 5 . try to check weather all the data is aviable inside your table or not

### 6 . try to group all the data with COL1 AND count occurences of each and every record based on col1 value

```python
query = "SELECT buying, count(*) FROM car_db.car GROUP BY buying"
cursor.execute(query)
cursor.fetchall()
```

## 7 . Try to filter a record where col 3 value will be 4 .

```python
query = "SELECT * FROM car_db.car WHERE doors = '4'"
cursor.execute(query)
cursor.fetchall()
```

## 8.  Try to update a col 3 value with 8 whereever you have value equal to 2

```python
query = "SELECT * FROM car_db.car WHERE doors = '8'"
cursor.execute(query)
cursor.fetchall()
```
