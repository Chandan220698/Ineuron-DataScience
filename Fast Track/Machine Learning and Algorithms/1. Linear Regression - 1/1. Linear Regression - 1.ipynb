```python
## First model between Sales vs TV
```

```python
df
```

```python
x = df[["TV"]]
x
```

```python
y = df['sales']
y
```

```python
# our eq:
# sales = m*TV + c
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear = LinearRegression()
```

```python
linear.fit(x, y)   # Calling the training find out best m and c
```

```python
linear.intercept_  # Value of C 
```

```python
linear.coef_      # slope m
```

```python
# Equation becomes:
# sales = 0.04753664*TV + 7.032593549127693   # This is our model
```

```python
file = "linear_reg.sav"   # sav bytecode format
```

```python
pickle.dump(linear, open(file, 'wb'))
```

```python
linear.predict([[45]])
```
