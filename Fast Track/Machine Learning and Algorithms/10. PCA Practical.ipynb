```python
## Step 1
## Condition Mean = 0 and SD = 1 not meet
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
scaled_data
```

```python
df = pd.DataFrame(scaled_data)
```

```python
df
```

```python
df.describe()
```

```python
## Objective acheived mean = 0 and SD = 1
```

```python
## Step 2: Find out Covariance Matrix
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit_transform(df)
```

```python
## 9 PCA for each column
pd.DataFrame(pca.fit_transform(df))  
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
```

```python
cusum = 0
for i in pca.explained_variance_ratio_:
    cusum += i
    print(cusum)
```

```python
pca1 = PCA(n_components=5)
pca1_data = pca1.fit_transform(df)
pca1_data
```

```python
x = pd.DataFrame(pca1_data, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5'])
x
```
