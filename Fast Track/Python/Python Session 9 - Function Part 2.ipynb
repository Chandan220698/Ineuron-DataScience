### Generator

```python
def test8(n):
    l = []
    for i in range(n):
        l.append(i**3)
        
    return l
```

```python
test8(6)
```

```python
## test8(600000000) ## This take so much time and consume all the CPU usage
```

```python
## To fix this we use generator
```

```python
def test9(n):
    for i in range(n):
        yield i**3
```

```python
test9(30)
```

```python
for i in test9(30):
    print(i)
```

```python
def genfib(n):
    a = 1
    b = 1
    output = []
    
    for i in range(n):
        output.append(a)
        a, b = b, a+b
        
    return output
```

```python
genfib(10)
```

```python
def genfib_yield(n):
    a = 1
    b = 1
    output = []
    
    for i in range(n):
        yield a
        a, b = b, a+b
```

```python
genfib_yield(6)
```

```python
for i in genfib_yield(6):
    print(i)
```

```python
for i in genfib_yield(600):
    print(i)
```
