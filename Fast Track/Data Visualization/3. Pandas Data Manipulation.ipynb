```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randn(10, 4), index = pd.date_range('2021-01-01', periods=10), columns=['a', 'b', 'c', 'd'])
```

```python
df
```

```python
df['new'] = df['c'] + df['d']
```

```python
df
```

```python
df['square_new'] = df['new'].apply(lambda x: x**2)
df
```

```python
df['new_log_d'] = df['d'].apply(lambda x: np.log(x))
df
```

```python
def test(x):
    if x < 0:
        return x
    else:
        return np.log(x)
```

```python
df['new_log_d-2'] = df['d'].apply(test)
df
```

```python
def test1(x):
    if x < 0:
        return 1
    else:
        return 0
```

```python
df['new_a'] = df['a'].apply(test1)
df
```
